```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale
```

```python
wine=pd.read_csv("wine.csv")
wine
```

```python
wine.info()
```

```python
wine.duplicated().value_counts()
```

```python
wine['Type'].value_counts()
```

```python
wine_1=wine.drop("Type", axis=True)
wine_1
```

```python
wine_ary=wine_1.values
wine_ary
```

```python
wine_norm=scale(wine_ary)
wine_norm
```

# PCA

```python
pca=PCA()
wine_pca=pca.fit_transform(wine_norm)
wine_pca
```

```python
pca=PCA(n_components = 10)
wine_pca=pca.fit_transform(wine_norm)
wine_pca
```

```python
pca.components_
```

```python
var=pca.explained_variance_ratio_
var
```

```python
var1=np.cumsum(np.round(var,4)*100)
var1
```

```python
plt.plot(var1,color='red')
```

# Final data frame after dimensional reduction

```python
final_df=pd.concat([wine['Type'],pd.DataFrame(wine_pca[:,0:3],columns=['PC1','PC2','PC3'])],axis=1)
final_df
```

```python
fig=plt.figure(figsize=(10,7))
sns.scatterplot(data=wine_1)
```

```python
fig=plt.figure(figsize=(10,7))
sns.scatterplot(data=final_df)
```

# Lets check with other clustering Algorithms

# Hierarchical Clustering

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import normalize
```

```python
plt.figure(figsize=(10,7))
dendrogram=sch.dendrogram(sch.linkage(wine_norm,'complete'))
```

```python
hclusters=AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='ward')
hclusters
```

```python
y=pd.DataFrame(hclusters.fit_predict(wine_norm),columns=['H_clusters'])
y['H_clusters'].value_counts()
```

```python
wine2=wine.copy()
wine2['H_cluster']=hclusters.labels_
wine2
```

```python
wine2.groupby('H_cluster').agg(['mean']).reset_index()
```

```python
from sklearn.metrics import silhouette_score
Hr_score = silhouette_score(wine_norm, hclusters.labels_, metric='euclidean')
Hr_score
```

```python
## silhouette_score is not good as optimal number of clusters are 3
```

# K-Means Clustering

```python
from sklearn.cluster import KMeans
```

```python
wcss=[]
for i in range (1,6):
    kmeans=KMeans(n_clusters=i,random_state=123)
    kmeans.fit(wine_norm)
    wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1,6),wcss)
plt.title('Elbow Graph')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
## Based on knee plot the optimal number of clusters are 3
```

```python
k_clusters=KMeans(3,random_state=123).fit(wine_norm)
k_clusters
```

```python
k_clusters.labels_
```

```python
wine2["K_cluster"]=k_clusters.labels_
wine2
```

```python
wine2.groupby('K_cluster').agg(['mean']).reset_index()
```

```python
from sklearn.metrics import silhouette_score
K_score = silhouette_score(wine_norm, k_clusters.labels_, metric='euclidean')
K_score
```

```python
## Both H_Clustering and K-Means clustering giving almost similar clusters(Based on silhouette_score)
```
